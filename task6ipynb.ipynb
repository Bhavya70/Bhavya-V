IMPORTING REQUIRED DEPENDENCIES


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import sklearn.datasets as dataset

```

IMPORTING DATASET

```python

iris=dataset.load_iris()

# Forming the iris dataframe
dataset=pd.DataFrame(iris.data, columns=iris.feature_names)

```

VISUALIZING THE ATTRIBUTES OF DATASET

```python


print(dataset.head(5))

```

```python
print(dataset.tail())
```

CHECKING WEATHER ANY MISSING DATA EXISTS IN DATASET




```python
dataset.isnull().sum()
```

```python
# IT IS CONCLUDED THAT THERE IS NO ANY MISSING DATA IN THE DATASET
```

```python
x = dataset
y = iris.target
```

SPLITTING THE DATASET FOR TRAINING AND TESTING

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

BUILDING OUR MODEL

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
# Install required libraries
!pip install pydotplus
!apt-get install graphviz -y
```

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(model, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

FINDING TRAINING ACCURACY

```python
from sklearn.metrics import accuracy_score
```

```python
prediction = model.predict(x_train)
```

```python
training_accuracy = accuracy_score(y_train, prediction)
```

```python
print("training accuracy is:",training_accuracy)
```

TRAINING ACCURACY :100%


LET US KNOW THE TESTING ACCURACY

```python
test_prediction = model.predict(x_test)
```

```python
test_accuracy = accuracy_score(y_test, test_prediction)
```

```python
print("testing accuracy is:", test_accuracy)
```

TESTING ACCURACY 93%

BUILDING A PREDICTIVE MODEL TO FIND THE CORRECT SPECIES WHEN WE FEED NEW DATA

```python
input_data = [6.8, 3.4, 5.8, 2.8]
input_data_to_numpy = np.asarray(input_data)
reshape_input_data = input_data_to_numpy.reshape(1,-1)
prediction = model.predict(reshape_input_data)


```

```python
print(prediction)
```

FROM THIS IT IS CONCLUDED WHEN WE FEED NEW DATA THIS MODEL CAN EASILY IDENTIFY THE CORRECT SPECIES TO WHICH IT BELONGS TO .
